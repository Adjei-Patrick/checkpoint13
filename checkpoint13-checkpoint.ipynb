```python
#ecrire un programme python pour creer un compte Ã  rebours
import time as tm

t = int(input("veiller saisir le temps en  seconde(s) svp: "))

def countdown(t):
    while t >= 0:
        mins,secs = divmod(t,60)
        temps = '{:02d}:{:02d}'.format(mins, secs) 
        print(temps, end="\r") 
        tm.sleep(1) 
        t -= 1
    print("le feu dans le trou !!")

countdown(t)
```
